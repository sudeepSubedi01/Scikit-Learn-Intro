```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_openml
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import confusion_matrix
```

```python
mnist = fetch_openml('mnist_784')
```

```python
X = pd.DataFrame(mnist['data'])
y = pd.DataFrame(mnist['target'])
print(type(X))
x_train,x_test,y_train,y_test = X[:60000],X[60000:],y[:60000],y[60000:]
# print(x_train.shape)
# print((x_train.loc[0]).to_numpy())
```

```python
# digit = x_train.loc[[4000],:].to_numpy()
# print(type(digit))
# print(digit.shape)
# image = digit.reshape((28,28))
# plt.scatter(image)
# digit_image = digit.reshape((28,28))
# plt.imshow(digit_image,cmap=matplotlib.cm.binary,interpolation='nearest')
# plt.axis('off')
print(y_test.head(10))
```

```python
x_train = (x_train/256)
x_test = (x_test/256)
# print(x_train.loc[[5000]].to_numpy())
model = MLPClassifier(solver = 'adam', activation='relu', hidden_layer_sizes=(64,64))
model.fit(x_train,y_train)
prediction = model.predict(x_test)
print(prediction)
```

```python
print(type(prediction))
print(prediction[:10])
```

```python
acc = confusion_matrix(y_test,prediction)
print(acc)
total_correct_predictions = acc.trace()
print('Total correct predictions:', total_correct_predictions)
total_predictions = acc.sum()
print('Total predictions: ', total_predictions)
accuracy = (total_correct_predictions / total_predictions) *100
print('The accuracy: ', accuracy)
```
