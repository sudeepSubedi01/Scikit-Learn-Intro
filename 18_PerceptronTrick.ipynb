```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from matplotlib.animation import FuncAnimation
import matplotlib.animation as animation
from sklearn.linear_model import LogisticRegression
```

```python
#creating dataset
X,y = make_classification(n_samples=100, n_features=2, n_informative=1,n_redundant=0, n_classes=2, n_clusters_per_class=1, random_state=41, hypercube=False,class_sep=10)
# print(X)
# print(y)
print(type(X))
print(X.shape)
print(y.shape)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X[:,0],X[:,1], c=y, cmap='winter',s=100)
```

```python
def step(z):
  return 1 if z>0 else 0
```

```python
def perceptron(X,y):
  X = np.insert(X,0,1,axis=1)   #inserting X0 column in the dataset
  weights = np.ones(X.shape[1])   #making initial weight
  lr = 0.1

  for i in range(1000):
    j = np.random.randint(0,100)      #random row number
    y_hat = step(np.dot(X[j],weights))    #product of matrix W and X
    weights = weights + lr * (y[j] - y_hat) * X[j]    # upadting weight
  return weights[0],weights[1:]     # returning W0 and W1,W2
```

```python
# print(X.shape)
# print(np.ones(X.shape[1]))
# print((X[10].shape))
# step(-0.5)
# np.random.randint(0,100)
```

```python
bias,coeff = perceptron(X,y)
print(bias)    # intercept = W0
print(coeff)      #coeff = W1, w2
```

```python
# general equation: Ax + By + C = 0
m = - (coeff[0] / coeff[1])    # slope of line = -A/B or W1/W2
y_intercept = -(bias / coeff[1])       # -C/B or W0/W1
```

```python
x_input = np.linspace(-3,3,100)
y_input = m * x_input + y_intercept
```

```python
plt.figure(figsize=(10,6))
plt.plot(x_input,y_input,color = 'red', linewidth=3)
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter', s=100)
plt.ylim(-3,2)
```
