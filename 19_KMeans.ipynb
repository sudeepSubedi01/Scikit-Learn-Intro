```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Mall_Customers.csv')
data.head()
# data.shape
```

```python
# visualizing the data that how many clusters can be formed visually
X1 = data.iloc[:,3]    # both are inputs
X2 = data.iloc[:,4]
# print(X.values)
# print(y)
plt.scatter(X1,X2)
```

```python
X = data.iloc[:,3:].values    # no inputs in kmeans
# print(X)
```

```python
# knowing how many clusters to form
wcss = []
for i in range(1,11):
  km = KMeans(n_clusters=i)
  km.fit(X)
  wcss.append(km.inertia_)
plt.plot(range(1,11), wcss)
```

```python
km = KMeans(n_clusters=5)
y_means = km.fit_predict(X)
print(y_means)
```

```python
plt.scatter(X[y_means == 0,0], X[y_means == 0,1], color = 'red')
plt.scatter(X[y_means == 1,0], X[y_means == 1,1], color='green' )
plt.scatter(X[y_means == 2,0], X[y_means == 2,1], color='blue' )
plt.scatter(X[y_means == 3,0], X[y_means == 3,1], color='yellow' )
plt.scatter(X[y_means == 4,0], X[y_means == 4,1], color='black' )
```
